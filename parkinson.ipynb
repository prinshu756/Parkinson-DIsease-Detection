```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn import svm 
from sklearn.preprocessing import StandardScaler
```

```python
import torch
```

```python
parkinson_data = pd.read_csv('Parkinsson disease.csv')
parkinson_data.tail()
```

```python
parkinson_data.shape
parkinson_data.info()
```

1 means the person has parkinson disease (positive) 
0 means the person is healthy (negative)

```python
X = parkinson_data.drop(columns=['name' , 'status'], axis =1)
Y = parkinson_data['status']
```

```python
print(X)
```

```python
print(Y)
```

Data preprocessing

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size=0.2 , random_state=1)
```

Standarizing the data

```python
scalar = StandardScaler()
scalar.fit(X_train)
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)
```

Training model using Support Vector Machine

```python
model = svm.SVC(kernel = 'linear')

#training the SVM model with training data

model.fit(X_train , Y_train)
X_train_prediction = model.predict(X_train)
accuracy_score = model.score(X_train , Y_train)

#getting the accuracy score of training data

print(accuracy_score)
```

```python
#training the SVM model with testing data

model.fit(X_test , Y_test)
X_test_prediction = model.predict(X_test)
accuracy_score = model.score(X_test , Y_test)

#getting the accuracy score of testing data

print(accuracy_score)
```

Building a prediction system for parkinson disease 

```python
input_data = (116.682,131.111,111.555,0.0105,0.00009,0.00544,0.00781,0.01633,0.05233,0.482,0.02757,0.03858,0.0359,0.0827,0.01309,20.651,0.429895,0.825288,-4.443179,0.311173,2.342259,0.332634
)
#changing the input data to numpy array 

data_np_arr = np.asarray(input_data)

#reshaping the numpy array as we are predicting for one instance

data_reshaped = data_np_arr.reshape(1, -1)
std_data = scalar.transform(data_reshaped)
prediction = model.predict(std_data)
print(prediction)

if (prediction[0] == 1) :
    print("Person have parkinson disease")
else:
    print("Person is healthy and not have Parkinson disease")
```
